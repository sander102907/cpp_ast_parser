```python
import os
from tqdm import tqdm
import math
import threading
```

```python
def preprocess(files, pbar):
    for file in files:
        with open(file) as program:
            to_remove = []
            code = program.read()
            for elements in [line.split(';') for line in code.split('\n')]:
                for element in elements:
                    if element.startswith('#include'):
                        to_remove.append(element)
                    if element.startswith('using namespace'):
                        to_remove.append(element + ';')

                    
            for el in to_remove:
                code = code.replace(el, '')

            output = open(f'{out_directory}{file.split("/")[-1]}', "w")
            output.write(code.strip())
            output.close()
            pbar.update()
```

```python
inp_directory = '../data/subset/cpp_preprocessed/'
out_directory =  '../data/subset/cpp_preprocessed2/'
amt_threads = 10

os.makedirs(out_directory, exist_ok=True) 

all_files = []

for _,_,files in os.walk(inp_directory):
    for file in files:
        if file.endswith('.cpp'):
            all_files.append(f'{inp_directory}{file}')

pbar = tqdm(total=len(all_files))

files_per_thread = math.ceil(
                len(all_files)/amt_threads)

threads = []

for index in range(0, len(all_files), files_per_thread):
    threads.append(
        threading.Thread(
            target=preprocess,
            args=(all_files[index: index + files_per_thread], pbar,)
            )
            )

for t in threads:
    t.start()

for t in threads:
    t.join()
            
```
